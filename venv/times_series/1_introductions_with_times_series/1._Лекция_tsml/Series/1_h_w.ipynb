```python
import pandas as pd
df_1, df_2, df_3, df_4, df_5, df_6 = \
    pd.read_csv("monthly-sales-of-company-x-jan-6.csv"), \
    pd.read_csv("monthly-boston-armed-robberies-j.csv"), \
    pd.read_csv("international-airline-passengers.csv"), \
    pd.read_csv("mean-monthly-air-temperature-deg.csv"), \
    pd.read_csv("weekly-closings-of-the-dowjones-.csv"), \
    pd.read_csv("daily-total-female-births-in-cal.csv")
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
all_series = {
    "Monthly sales of company X": df_1["Count"],
    "Monthly Boston armed robberies": df_2["Count"],
    "International airline passengers: monthly totals in thousands": df_3["Count"],
    "Mean monthly air temperature (Deg. F) Nottingham Castle": df_4["Deg"],
    "Weekly closings of the Dow-Jones industrial average": df_5["Close"],
    "Daily total female births in California": df_6["Count"]
}
```

```python
with plt.style.context('bmh'):
    plt.figure(figsize=(16, 8))
    layout = (3, 2)
    for i, key in enumerate(all_series.keys()):
        x = i % 2
        y = int((i - x) / 2)
        
        ts_ax = plt.subplot2grid(layout, (y, x))
        all_series[key].plot(ax=ts_ax, color='blue')
        ts_ax.set_title(key)
        
    plt.tight_layout()
```
